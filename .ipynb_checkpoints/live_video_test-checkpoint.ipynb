```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the model and initialize face detector
model = load_model('model_file.h5')
faceDetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
labels_dict = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Neutral', 5: 'Sad', 6: 'Surprise'}
```

```python
# Initialize video capture
video = cv2.VideoCapture(0)

while True:
    ret, frame = video.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceDetect.detectMultiScale(gray, 1.3, 3)
    for x, y, w, h in faces:
        sub_face_img = gray[y:y+h, x:x+w]
        resized = cv2.resize(sub_face_img, (48, 48))
        normalize = resized / 255.0
        reshaped = np.reshape(normalize, (1, 48, 48, 1))
        result = model.predict(reshaped)
        label = np.argmax(result, axis=1)[0]
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 1)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (50, 50, 255), 2)
        cv2.rectangle(frame, (x, y-40), (x+w, y), (50, 50, 255), -1)
        cv2.putText(frame, labels_dict[label], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    
    cv2.imshow("Frame", frame)
    if cv2.waitKey(1) == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import time

# Load the model and initialize face detector
model = load_model('model_file.h5')
faceDetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
labels_dict = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Neutral', 5: 'Sad', 6: 'Surprise'}

# Initialize video capture
video = cv2.VideoCapture(0)

# Dictionary to store emotion scores and a list to keep track of timestamps
emotion_scores = {key: [] for key in labels_dict.keys()}
start_time = time.time()

while True:
    ret, frame = video.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceDetect.detectMultiScale(gray, 1.3, 3)
    for x, y, w, h in faces:
        sub_face_img = gray[y:y+h, x:x+w]
        resized = cv2.resize(sub_face_img, (48, 48))
        normalize = resized / 255.0
        reshaped = np.reshape(normalize, (1, 48, 48, 1))
        result = model.predict(reshaped)
        emotion_probabilities = result[0]
        
        for i in range(len(labels_dict)):
            emotion_scores[i].append(emotion_probabilities[i])

        label = np.argmax(result, axis=1)[0]
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 1)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (50, 50, 255), 2)
        cv2.rectangle(frame, (x, y-40), (x+w, y), (50, 50, 255), -1)
        cv2.putText(frame, f"{labels_dict[label]}: {emotion_probabilities[label]:.2f}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    
    cv2.imshow("Frame", frame)

    # Calculate and display average emotion scores every minute
    current_time = time.time()
    if current_time - start_time >= 60:
        avg_emotion_scores = {labels_dict[key]: np.mean(emotion_scores[key]) if emotion_scores[key] else 0 for key in emotion_scores}
        print("Average Emotion Scores over the last minute:")
        for emotion, score in avg_emotion_scores.items():
            print(f"{emotion}: {score:.2f}")
        # Reset the data for the next minute
        emotion_scores = {key: [] for key in labels_dict.keys()}
        start_time = current_time

    if cv2.waitKey(1) == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
